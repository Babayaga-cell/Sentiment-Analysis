## Run All Necessary Imports

```python
import re
import csv
import string
import pandas as pd

from langdetect import detect        #Language Detection

from googletrans import Translator   #Google Translator
translator = Translator()

from textblob import TextBlob

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer   #VADER Sentiment-Lexicon Dictionary
analyzer = SentimentIntensityAnalyzer()
```

## Read Dataset as CSV file

```python
#100 is the middle ground environment of the Google Translator API current version 4.0.0rc1

Dataset= pd.read_csv('5k_shuffle_dataset201.csv')

Dataset.head()
```

## Dataset Current Size

```python
print('Dataset size:',Dataset.shape)
```

## Data Cleaning Part 1

```python
def cleanTxt(text):
    text = re.sub(r'@[A-Za-z0-9]+','',text)    #Removes of @mentions
    text = re.sub(r'#','', text)               #Removes of the '#' symbol
    text = re.sub(r'RT[\s]+','', text)         #Removes of RT
    text = re.sub(r'https?:\/\/\S+', '',text)  #Removes the hyperlink
    
    return text

cleantTxt = Dataset['Tweets'].apply(cleanTxt)

Dataset['cleantTxt'] = cleantTxt

Dataset.head()
```

## Data Cleaning Part 2

```python
def cleaning_PicURL (text):
    text = re.sub(r'pic.twitter.com/[\w]*',"", text)   #Removes picURLs
    return text

cleaningPicUrl = Dataset['cleantTxt'].apply(lambda x: cleaning_PicURL(x))


Dataset['cleaningPicUrl'] = cleaningPicUrl\

Dataset.head()
```

## Data Cleaning Part 3

```python
Dataset['text_lower']  = Dataset['cleaningPicUrl'].str.lower()  ##Lowers all capitalized letters

Dataset.head()
```

## Data Cleaning Part 4

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])   #Removes punctuations
    text = re.sub('[0-9]+', '', text)
    return text

Dataset['remov_punc'] = Dataset['text_lower'].apply(lambda x: remove_punct(x))


Dataset.head()
```

## Data Cleaning Part 5

```python
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # Removes emoticons
                           u"\U0001F300-\U0001F5FF"  # Removes symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # Removes transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # Removes flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)


Dataset['rem_emoji'] = Dataset['remov_punc'].apply(lambda x: remove_emoji(x))
Dataset.head()
```

## Detect the Languages per Tweets

```python
Dataset['Language'] = Dataset['rem_emoji'].apply(detect)
Dataset.head()
```

## Detect and Display Tweets that Using 'EN' or 'TL' Languages.

```python
def detect_lang(text):                                                #Detects Tweets that uses 'EN' or 'TL'
    try:
        return detect(text) == 'en' or detect(text) == 'tl'
    except:
        return False
    
Dataset = Dataset[Dataset['rem_emoji'].apply(detect_lang)]
Dataset.head()
```

## Translate Detected 'TL' Tweets to English (or 'EN') and remain already 'EN' Tweets.

```python
Dataset['Translated']=Dataset['rem_emoji'].apply(lambda x: translator.translate(x, src='auto', dest='en').text )
Dataset.head()
```

## Determine the Polarity of the Translated Tweets and the original English Tweets.

```python
Dataset['Polarities'] = Dataset['Translated'].apply(lambda Translated:analyzer.polarity_scores(Translated))
Dataset.head()
```

## Dataset New Size

```python
print('Dataset size:',Dataset.shape)
```

## Export the CSV with the addition of Translated and Polarities columns

```python
Dataset.to_csv('5k_shuffle_dataset_part1.csv',index=False)
```
